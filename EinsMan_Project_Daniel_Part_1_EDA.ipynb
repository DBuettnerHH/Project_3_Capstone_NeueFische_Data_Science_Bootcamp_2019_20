## Project Description:

### Quadra Energy EinsMan Loss Work Predictions

Power production from renewable energy sources, like wind or solar radiation, are very volatile due to changing weather conditions. The highly variable weather characteristics of Germany’s mid-latitude climate and its uncertain forecasts make the anticipation of the actual instantaneous overall power supply a tough call for all market players.<br><br>
The strong development of on- and off-shore wind power of the past 5-10 years in Germany (and to some extent solar power as well) and the simultaneously slow grid expansion led to vastly increased occurrences of grid bottleneck situations (oversupply, missing powerline capacities), especially during strong wind conditions. In order to maintain the overall stability of the power grid, the operators have actions to be taken. After all other existing measures are exhausted, grid operators can temporarily exclude wind farms and their respective energy supply from the grid in the necessary order of magnitude.<br><br>
These exclusions are called feed-in management (in German “Einspeisemanagement”), or in short Einsman/Eisman. The energy “lost” during these Einsman situations are called loss work. The total loss work for Germany amounted to 5,5 TWh in 2017, an increase by a factor of 75 compared to 2009.<br><br>
Anticipating these energy management situations in advance is crucial in order to prevent direct marketers of renewable energy from having to fill supply gaps by buying on the intraday exchange to often (highly) increased prices compared to preceding day ahead prices.<br><br>
For the purpose of deriving such anticipation/prediction of these Einsman regulation events, Quadra Energy provided the following data about power generation and meteorological model features:

## Data Sets:

### Quadra Energy Data Description: 

- Data from Wind Power Plants in Schleswig-Holstein (summed up production values, normed)
- Covering around 300 MW of installed power
- Time Span: 2018-01-01 00:00:00 to 2019-10-31 23:40:00
- Time Resolution: 10 min
<br><br>
- Variables:
    
| Variable | Description |
| :-------- | :--------- |
| power_available_norm | potentially available power supply (power curves * wind conditions), normed: available power supply / installed power |
| losses_rtu_einsman_norm | Loss work due to EinsMan events, power supply shut down during management situations, normed: loss work / installed power|

### Numerical Weather Prediction, GFS (Global Forecast System, NCEP) Model Data:

- Data from GFS grid cells over the target region of Schleswig-Holstein
- Mean Values over all selected grid cells
- Time Span: 2018-01-01 06:00:00 to 2019-08-26 18:00:00

### Upsample GFS Data to Quadra Data resolution (10 min):

- Resample GFS data to 10 min frequency.
- GFS Data are hourly instantaneous values
- Wind Direction Variables will be allocated since simply calculating a mean can result in erorrneous values (e.g. (350+10)/2 = 180):
    - 'wdir_10m_dn', 'wdir_100m_dn', 'wdir_pfg30_dn', 'wdir_pl925_dn'
        - forward (forward fill 3 steps) and
        - backward (backward fill 2 steps)
        
- All other Meteorological Variables will be linearly interpolated between full hours
    - otherwise interpolation could result in values greater or smaller than two consecutive 1h values between which the interpolation is applied
    
- Except for 'Elevation', which is interpolated 'quadratic' (giving better representation during midday/midnight turnover points)

```python
# Resample GFS data to 10 min frequency:

df_gfs_10min = df_gfs_1h.resample('10T').asfreq()
```

```python
df_gfs_10min.head(10)
```

#### Interpolate Variables:

```python
# Allocate hourly GFS Data to 10 min values --> ffill 3 steps, bfill 2 steps

cols_wdir = ['wdir_10m_dn', 'wdir_100m_dn', 'wdir_pfg30_dn', 'wdir_pl925_dn']

# Allocating Wind Direction data (as interpolation won't work for all directions), forward/backward filling: 
df_gfs_10min[cols_wdir] = df_gfs_10min[cols_wdir].fillna(method='ffill', limit=3)
df_gfs_10min[cols_wdir] = df_gfs_10min[cols_wdir].fillna(method='bfill', limit=2)

# Allocating Creation dates:
df_gfs_10min['created_on'] = df_gfs_10min['created_on'].fillna(method='ffill', limit=5)

# Interpolating Elevation data in quadratic form (better midday/midnight representation than 1h res):
df_gfs_10min['elevation'] = df_gfs_10min['elevation'].interpolate(method='quadratic')


# Interpolating all remaining variables in linear form:
# (otherwise interpolated values could result which are greater/smaller than two consecutive 1h values)
df_gfs_10min = df_gfs_10min.interpolate(method='linear')
```

### Monthly Sums EinsMan:

```python
# Build DataFrame Quadra Monthly Sums:
# calculation of MWh: data represents around 280 MW of total installed power
#                     data points are 10 min resolution --> multiply each data point by 1/6 for sums to
#                                                           represent accumulated hours

# EinsMan_norm, EinsMan_MWh,
# Power_avail_norm, Power_avail_MWh
# Power_actual_norm (Power_avail - EinsMan), Power_actual_MWh
# Power_max_norm (count of data points), Power_max_MWh
# EinsMan_frac_avail (fraction of available power suppy cut off due to EinsMan)
# EinsMan_frac_max (fraction of max power supply cut off due to EinsMan)
# Power_avail_frac_max (fraction of max power supply potentially available)
# Power_actual_frac_max (fraction of max power supply actually produced/supplied)


df_quadra_mon_sum = df_gq_10min[['losses_rtu_einsman_norm', 'power_available_norm']].resample('M').sum()
df_quadra_mon_sum.rename(columns={'losses_rtu_einsman_norm': 'EinsMan_norm',
                                       'power_available_norm': 'Power_avail_norm'}, inplace=True)
df_quadra_mon_sum.index = df_quadra_mon_sum.index.to_period('M')

df_quadra_mon_sum['Power_actual_norm'] = df_quadra_mon_sum['Power_avail_norm']\
                                         - df_quadra_mon_sum['EinsMan_norm']

df_quadra_mon_sum['Power_max_norm'] = df_gq_10min.losses_rtu_einsman_norm.groupby(df_gq_10min.index.to_period('M')).agg('count')
```

```python
#df_quadra_mon_sum
```

```python
# Monthly sum in MWh:

df_quadra_mon_sum_MWh = df_quadra_mon_sum * 280 / 6
df_quadra_mon_sum_MWh.rename(columns={'EinsMan_norm': 'EinsMan_MWh',
                                      'Power_avail_norm': 'Power_avail_MWh',
                                      'Power_actual_norm': 'Power_actual_MWh',
                                      'Power_max_norm': 'Power_max_MWh'}, inplace=True)
```

```python
df_quadra_mon_sum_MWh['2018'].EinsMan_MWh.sum()
```

```python
# Monthly fractions:

df_quadra_mon_frac = df_quadra_mon_sum.EinsMan_norm / df_quadra_mon_sum.Power_avail_norm * 100
df_quadra_mon_frac = pd.DataFrame(data=df_quadra_mon_frac, 
                                  columns=['EinsMan_frac_avail'],
                                  index=df_quadra_mon_sum.index)
df_quadra_mon_frac['EinsMan_frac_max']        = df_quadra_mon_sum.EinsMan_norm\
                                                / df_quadra_mon_sum.Power_max_norm * 100
df_quadra_mon_frac['Power_avail_frac_max']    = df_quadra_mon_sum.Power_avail_norm\
                                                / df_quadra_mon_sum.Power_max_norm * 100
df_quadra_mon_frac['Power_actual_frac_max']   = df_quadra_mon_sum.Power_actual_norm\
                                                / df_quadra_mon_sum.Power_max_norm * 100
```

### Daily Sums EinsMan:

```python
# Build DataFrame Quadra Daily Sums:
# calculation of MWh: data represents around 280 MW of total installed power
#                     data points are 10 min resolution --> multiply each data point by 1/6 for sums to
#                                                           represent accumulated hours

# EinsMan_norm, EinsMan_MWh,
# Power_avail_norm, Power_avail_MWh
# Power_actual_norm (Power_avail - EinsMan), Power_actual_MWh
# Power_max_norm (count of data points), Power_max_MWh
# EinsMan_frac_avail (fraction of available power suppy cut off due to EinsMan)
# EinsMan_frac_max (fraction of max power supply cut off due to EinsMan)
# Power_avail_frac_max (fraction of max power supply potentially available)
# Power_actual_frac_max (fraction of max power supply actually produced/supplied)


df_quadra_day_sum = df_gq_10min[['losses_rtu_einsman_norm', 'power_available_norm']].resample('D').sum()
df_quadra_day_sum.rename(columns={'losses_rtu_einsman_norm': 'EinsMan_norm',
                                       'power_available_norm': 'Power_avail_norm'}, inplace=True)
df_quadra_day_sum.index = df_quadra_day_sum.index.to_period('D')

df_quadra_day_sum['Power_actual_norm'] = df_quadra_day_sum['Power_avail_norm']\
                                         - df_quadra_day_sum['EinsMan_norm']

df_quadra_day_sum['Power_max_norm'] = df_gq_10min.losses_rtu_einsman_norm.groupby(df_gq_10min.index.to_period('D')).agg('count')
```

```python
max(df_quadra_day_sum.EinsMan_norm)
```

```python
#df_quadra_day_sum[:10]
```

```python
# Daily sum in MWh:

df_quadra_day_sum_MWh = df_quadra_day_sum * 280 / 6
df_quadra_day_sum_MWh.rename(columns={'EinsMan_norm': 'EinsMan_MWh',
                                      'Power_avail_norm': 'Power_avail_MWh',
                                      'Power_actual_norm': 'Power_actual_MWh',
                                      'Power_max_norm': 'Power_max_MWh'}, inplace=True)
```

```python
#df_quadra_day_sum_MWh[:10]
```

```python
# Daily fractions:

df_quadra_day_frac = df_quadra_day_sum.EinsMan_norm / df_quadra_day_sum.Power_avail_norm * 100
df_quadra_day_frac = pd.DataFrame(data=df_quadra_day_frac, 
                                  columns=['EinsMan_frac_avail'],
                                  index=df_quadra_day_sum.index)
df_quadra_day_frac['EinsMan_frac_max']        = df_quadra_day_sum.EinsMan_norm\
                                                / df_quadra_day_sum.Power_max_norm * 100
df_quadra_day_frac['Power_avail_frac_max']    = df_quadra_day_sum.Power_avail_norm\
                                                / df_quadra_day_sum.Power_max_norm * 100
df_quadra_day_frac['Power_actual_frac_max']   = df_quadra_day_sum.Power_actual_norm\
                                                / df_quadra_day_sum.Power_max_norm * 100
```

```python
#df_quadra_day_frac[:10]
```

```python
# Combine daily sums (normed and MWh) and fractions to one DataFrame:

df_quadra_day_sum_frac = pd.concat([df_quadra_day_sum,
                                    df_quadra_day_sum_MWh,
                                    df_quadra_day_frac], axis=1)
```
